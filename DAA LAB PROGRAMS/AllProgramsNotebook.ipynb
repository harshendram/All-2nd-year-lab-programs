 Finding  Nth Factorial using recursive vs Non recursive  and Analyze time  

```python
import time as Time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def fact1(n):
    if n == 0:
        return 1
    else:
        return n * fact1(n-1)

def fact2(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

time1 = []
inputs = []
time2 = []
for i in range(10 ,500 ,10):
    inputs.append(i)
    start = Time.time()
    fact1(i)
    end = Time.time()
    time1.append(end - start)

    start = Time.time()
    fact2(i)
    end = Time.time()
    time2.append(end - start)

plt.plot(inputs, time1, label='Recursive')
plt.plot(inputs, time2, label='Non-Recursive')
plt.xlabel('Input Size')
plt.ylabel('Time')
plt.legend()
plt.show()
```

 Selection Sort vs Bubble Sort (TC Comparision)

```python
import matplotlib.pyplot as plt
import time
from random import randint

def bubblesort(a,n):
    for i in range(0,n-1):
        for j in range(0,n-1-i):
            if(a[j+1] < a[j]):
                a[j+1],a[j] = a[j],a[j+1]

def selectionsort(a,n):
    for i in range(0,n-1):
        min = i
        for j in range(i+1,n):
            if(a[j] < a[min]):
                min = j
        a[i],a[min] = a[min],a[i]

def main():
    time_array_bubb = []
    time_array_sele = []
    num_of_ele_array = []

    for n in range(10,1001,10):

        array1 = []
        num_of_ele_array.append(n)

        for i in range(0,n):
            array1.append(randint(1,n))
        array2 = array1

        start = time.time()
        bubblesort(array1,n)
        end = time.time()
        gaptime = end - start
        time_array_bubb.append(gaptime)

        start = time.time()
        selectionsort(array2,n)
        end = time.time()
        gaptime = end - start
        time_array_sele.append(gaptime)


    plt.plot(num_of_ele_array,time_array_bubb,label = 'bubble sort')
    plt.plot(num_of_ele_array,time_array_sele,label = "selection sort")
    plt.xlabel("Input size")
    plt.ylabel("Time")
    plt.legend()
    plt.show()
main()
```

 Linear Search vs Binary Search(TC comparision)


```python
import matplotlib.pyplot as plt
import time


def linearSearch(a,n,key):
    lcnt = 0
    for i in range(0,n):
        lcnt = lcnt + 1
        if(a[i] == key):
            print("key is found at the position=",i)
            return lcnt
    print("key is not found!")
    return lcnt

def binarySearch(a,n,key):
    low = 0
    high = n-1
    bcount = 0
    while low <= high:
        mid = (low+high) // 2
        bcount += 1
        if(a[mid] == key):
            print("key is found")
            return bcount
        elif(key < a[mid]):
            high = mid -1
        else:
            low = mid +1
    print("key not found")
    return bcount



def main():
    n = int(input("Enter n:"))
    a = []
    for i in range(n):
        a.append(i)
    #a = list(range(n))
    print(a)
    key = int(input("Enter key:"))
    start = time.time()
    Lcount  = linearSearch(a,n,key)
    end  = time.time()
    Lgaptime = end - start
    start = time.time()
    Bcount = binarySearch(a,n,key)
    end = time.time()
    Bgaptime = end - start

    print("Number of comparisons in Linear search = ",Lcount)
    print("Number of comparisons in Binary search = ",Bcount)

    plt.bar(['Linear search','Binary search'],[Lgaptime,Bgaptime],label = ['Linear search','Binary search'],color = ['red','green'])
    plt.xlabel("input size n:")
    plt.ylabel("Time sec")
    plt.legend()
    plt.show()

main()
```

Merge Sort Analysis

```python
import matplotlib.pyplot as plt
import time as t
import random

def divide(a, low, high):
    if low < high:
        mid = (high + low) // 2
        divide(a, low, mid)
        divide(a, mid + 1, high)
        merge(a, low, mid, high)

def merge(a, low, mid, high):
    temp = [0] * len(a)
    i = low
    j = mid + 1
    k = low

    while i <= mid and j <= high:
        if a[i] <= a[j]:
            temp[k] = a[i]
            i += 1
        else:
            temp[k] = a[j]
            j += 1
        k += 1

    while i <= mid:
        temp[k] = a[i]
        i += 1
        k += 1

    while j <= high:
        temp[k] = a[j]
        j += 1
        k += 1

    for k in range(low, high + 1):
        a[k] = temp[k]

if __name__ == "__main__":
    x = []
    y = []

    for n in range(1000, 10001, 1000):
        a = []
        x.append(n)
        for i in range(n):
            a.append(random.randint(1, n))

        start = t.time()
        divide(a, 0, len(a) - 1)
        end = t.time()

        gap = end - start
        y.append(gap)

    plt.plot(x, y)
    plt.xlabel("n size")
    plt.ylabel("Time")
    plt.title("Merge Sort Time Complexity")
    plt.grid(True)
    plt.show()
```

Finding even elements in a matrix--TC analysis

```python
import matplotlib.pyplot as plt
import time
import numpy as np

def find_even_elements(matrix):
    even_elements = []
    rows, cols = matrix.shape
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] % 2 == 0:
                even_elements.append(matrix[i][j])
    return even_elements

def main():
    num_of_ele_matrix = []
    time_array_matrix = []

    for n in range(100, 10000, 100):
        matrix = np.random.randint(1, 1000, size=(n, n))
        num_of_ele_matrix.append(n)

        start = time.time()
        find_even_elements(matrix)
        end = time.time()

        gaptime = end - start
        time_array_matrix.append(gaptime)

    plt.plot(num_of_ele_matrix, time_array_matrix, label='Even element in matrix')
    plt.xlabel("Input size (n)")
    plt.ylabel("Time (seconds)")
    plt.title("Even Elements in Matrix")
    plt.legend()
    plt.show()

main()
```

Krushkal ALgo

```python
p = []

def kruskal(graph, n):
    for i in range(n):
        p.append(i)
    sum = 0
    ecount = 0
    while ecount < n - 1:
        min = 999
        for i in range(n):
            for j in range(n):
                if i != j and graph[i][j] < min:
                    min = graph[i][j]
                    u = i
                    v = j
        if parent(u) != parent(v):
            print("connect", u, "--->", v, "=", graph[u][v])
            ecount += 1
            sum += graph[u][v]
            p[v] = u
        graph[u][v] = graph[v][u] = 999
    print("MST Total Cost=", sum)

def parent(x):
    while x != p[x]:
        x = p[x]
    return x

def main():
    graph = [
        [0, 30, 40, 999, 999, 999],
        [30, 0, 50, 10, 999, 999],
        [40, 50, 0, 999, 20, 60],
        [999, 10, 999, 0, 60, 70],
        [999, 999, 20, 60, 0, 80],
        [999, 999, 999, 70, 80, 0]
    ]
    n = len(graph[0])

    print(".......weight matrix........")
    for row in graph:
        print(row)

    print("...........MST Edge..........")
    kruskal(graph, n)

main()
```

Floyd Algorithm

```python
def floyd(dist,n):
    for k in range(n):
        for i in range(n):
            for j in range(n):
                 dist[i][j]=min(dist[i][j],(dist[i][k]+dist[k][j]))
def main():
    dist=[ [0,999,3,999],
        [2,0,999,999],
        [999,7,0,1],
        [6,999,999,0]]
    n=len(dist[0])
    print(".....weight matrix........")
    print(dist)
    floyd(dist,n)
    print("....All pair shortest path....")
    print(dist)
main()
```

```python
visited = []
dist = []

def dijkstra (graph,n,source):
    for i in range(n):
        visited.append(0)
        dist.append(graph[source][i] )
    visited[source] = 1
    #dist[ source ] = 0
    for ecount in range(n):
        min = 999  #infinite
        #now to find minimum
        for i in range(n):
            if(visited[i] == 0 and dist[i] < min):   #not visited and dist should be less that min
                min = dist[i]
                u = i
        visited[u] = 1
        for v in range(n):
            if(visited[v] == 0 and dist[u]+graph[u][v] < dist[v]):
                dist[v] = graph[u][v] + dist[u]

def main():
    graph = [[0,3,999,7,999],
            [3,0,4,2,999],
            [999,4,0,5,6],
            [7,2,5,0,4],
            [999,999,6,4,0]]
    n = 5
    source = int(input("Enter the source:"))
    print("---------------dist matrix--------------")
    for row in graph:
        print("       ",row)
    dijkstra(graph,n,source)
    print("Shortest path from source =",source)
    for i in range(n):
        print(source,"----->",i,"====",dist[i])

main()
```

Knapsack - partial or fractional version



```python
def fractional_knapsack(W, weights, values):
    n = len(values)

    items = [(values[i] / weights[i], weights[i], values[i]) for i in range(n)]
    items.sort(reverse=True)

    total_value = 0.0

    for ratio, weight, value in items:
        if W == 0:
            break
        if weight <= W:

            total_value += value
            W -= weight
        else:

            total_value += ratio * W
            W = 0

    return total_value

# Example usage
W = 50
weights = [10, 20, 30]
values = [60, 100, 120]

print("Maximum value in fractional knapsack:", fractional_knapsack(W, weights, values))
```

Fibonacci with memoization

```python
def fib_memo(n, memo):
    if memo[n] != -1:
        return memo[n]
    if n <= 1:
        memo[n] = n
    else:
        memo[n] = fib_memo(n - 1, memo) + fib_memo(n - 2, memo)
    return memo[n]

n = int(input("Enter number of terms: "))
memo = [-1] * (n)
for i in range(n):
    print(fib_memo(i, memo), end=" ")
```

Factorial with memoization

```python
def factorial_memo(n, memo):
    if memo[n] != -1:
        return memo[n]
    if n == 0 or n == 1:
        memo[n] = 1
    else:
        memo[n] = n * factorial_memo(n - 1, memo)
    return memo[n]

n = int(input("Enter a number: "))
memo = [-1] * (n + 1)
print(f"Factorial of {n} is {factorial_memo(n, memo)}")
```

Exercise 4.1: Insertion Sort

Implement the insertion sort algorithm, which uses the decrease-and-conquer technique, to sort a list of numbers.

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Example usage:
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)
```

Exercise 4.4: Binary Search
Use the reduce-and-conquer technique to implement the binary search algorithm on a sorted list of integers. The algorithm should return the index of the target element if found, or -1 if not found.

```python
def binary_search(arr, target, low, high):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif target < arr[mid]:
        return binary_search(arr, target, low, mid - 1)
    else:
        return binary_search(arr, target, mid + 1, high)

# Example usage:
arr = [2, 4, 7, 10, 15, 20, 25]
target = 15
index = binary_search(arr, target, 0, len(arr) - 1)

if index != -1:
    print(f"Element {target} found at index {index}")
else:
    print(f"Element {target} not found")
```

Problem Statement (Exercise 5.1):

Design a divide-and-conquer algorithm to find the maximum element in an array of n integers. Your algorithm should recursively divide the array into halves and return the maximum of the two halves.

```python
def find_max(arr, low, high):
    if low == high:
        return arr[low]
    mid = (low + high) // 2
    left_max = find_max(arr, low, mid)
    right_max = find_max(arr, mid + 1, high)
    return max(left_max, right_max)

# Example usage:
arr = [12, 45, 23, 67, 34, 89, 10]
print("Maximum element is:", find_max(arr, 0, len(arr) - 1))
```

Problem Statement (Exercise 5.3):

Design a divide-and-conquer algorithm to compute the sum of all elements in an array of n integers by recursively dividing the array into halves and summing the results.

```python
def array_sum(arr, low, high):
    if low == high:
        return arr[low]
    mid = (low + high) // 2
    left_sum = array_sum(arr, low, mid)
    right_sum = array_sum(arr, mid + 1, high)
    return left_sum + right_sum

# Example usage:
arr = [2, 4, 6, 8, 10]
print("Sum of array elements:", array_sum(arr, 0, len(arr) - 1))
```
