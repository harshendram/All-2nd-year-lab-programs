```python
import time as Time
import matplotlib.pyplot as plt


def fact1(n):
    if n == 0:
        return 1
    else:
        return n * fact1(n-1)

def fact2(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result   

time1 = []
inputs = []
time2 = []
for i in range(10 , 1001,10):
    inputs.append(i)
    start = Time.time()
    fact1(i)
    end = Time.time()
    time1.append(end - start)
    
    start = Time.time()
    fact2(i)
    end = Time.time()
    time2.append(end - start)

plt.plot(inputs, time1, label='Recursive')
plt.plot(inputs, time2, label='Non-Recursive')
plt.xlabel('Input Size')
plt.ylabel('Time')
plt.legend()
plt.show()
```
